```python
import pandas as pd

# plotting:
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data = pd.read_csv("../../data/dydx.csv")
data["diff_price_btc"] = data["price_btc_math"] - data["price_btc_numeric"]
data["diff_price_eth"] = data["price_eth_math"] - data["price_eth_numeric"]
data
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))

ax.plot(
    data.dollar_amt_in,
    data.diff_price_btc,
    'b.',
    label="BTC"
)

ax.plot(
    data.dollar_amt_in,
    data.diff_price_eth,
    'r.',
    label="ETH"
)

ax.set_xlabel("Dollar amount in [$]", fontsize=20)
ax.set_ylabel("Price math - Price numeric [$]", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='x', which='major', colors='black')

ax.set_xscale("log")

ax.legend(fontsize=20)
```
